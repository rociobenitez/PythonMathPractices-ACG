**Datos:**

Un multiconjunto es un conjunto en el que un elemento puede repetirse, es decir, cada elemento posee una multiplicidad (un número natural) que indica cuántas veces el elemento es miembro del conjunto. Por ejemplo, en el multiconjunto `{a, a, b, b, b, c}`, las multiplicidades de los miembros a, b, y c son 2, 3, y 1, respectivamente.

Al igual que los conjuntos, poseen las siguientes características y operaciones:
- Cardinalidad: indica el número de elementos del multiconjunto. Por ejemplo, la cardinalidad del multiconjunto `{a, a, b, b, b, c}` es 6 (la suma de sus multiplicidades).
- Inserción: permite insertar una ocurrencia de un elemento en el multiconjunto.
- Eliminación: permite eliminar una ocurrencia de un elemento del multiconjunto.
- Comparación: compara dos multiconjuntos para determinar si son iguales.
- Pertenencia: determina si un elemento pertenece al multiconjunto.
- Subconjunto: determina si un multiconjunto es subconjunto de otro.
- Unión: conjunción de todos los elementos de dos multiconjuntos (sumando sus multiplicidades si un elementos está en los dos).
- Intersección: elementos que están en los dos multiconjuntos quedándonos con la multiplicidad más pequeña.
- Diferencia: restar a un multiconjunto los elementos de otro.

En la teoría de conjuntos, los multiconjuntos pueden representarse de diversas maneras, y cada una de estas representaciones ofrece ventajas únicas en términos de manipulación y acceso a los datos.

Una opción común en programación es el uso de **diccionarios**, donde cada clave corresponde a un elemento del multiconjunto y su valor asociado indica la multiplicidad o frecuencia de ese elemento. Esta forma es eficiente para operaciones como agregar, eliminar o consultar la frecuencia de un elemento. Otra representación es mediante **listas con repeticiones**, donde se incluyen todos los elementos, incluyendo sus repeticiones; aunque es más intuitiva, puede ser menos eficiente para listas grandes. Además, algunos lenguajes ofrecen estructuras especializadas como `Counter` en Python, que es un tipo de diccionario diseñado específicamente para contar ocurrencias.

```python
### TODO: Crear una función que dados dos multiconjuntos devuelva su intersección
def interseccion(multiconjunto1, multiconjunto2):

    frecuencias1 = {}
    frecuencias2 = {}
    multiconjunto_interseccion = []

    # Contar las frecuencias de cada elemento en multiconjunto1
    for elemento in multiconjunto1:
        if elemento in frecuencias1:
            frecuencias1[elemento] += 1
        else:
            frecuencias1[elemento] = 1

    # Contar las frecuencias de cada elemento en multiconjunto2
    for elemento in multiconjunto2:
        if elemento in frecuencias2:
            frecuencias2[elemento] += 1
        else:
            frecuencias2[elemento] = 1

    # Determinar la intersección
    for elemento in frecuencias1:
        if elemento in frecuencias2:
            veces_minimas = min(frecuencias1[elemento], frecuencias2[elemento])
            # Añadir múltiples elementos a la lista multiconjunto_interseccion
            multiconjunto_interseccion.extend([elemento] * veces_minimas)  

    return multiconjunto_interseccion

print(mc_lista)
print(multiconjunto([1,3,4]))
print(interseccion(multiconjunto([1,3,4]), mc_lista))
```

```python
### TODO: Crear una función que dados dos multiconjuntos devuelva su diferencia

from collections import Counter

def diferencia(multiconjunto1, multiconjunto2):
    contador1 = Counter(multiconjunto1)
    contador2 = Counter(multiconjunto2)

    # La diferencia de multiconjuntos se representa como la resta de contadores
    resultado = list((contador1 - contador2).elements())
    return resultado

# Ejemplo de uso de la función
print("Multiconjunto 1 (mc_lista):", mc_lista)
print("Multiconjunto 2:", multiconjunto([1, 2, 3, 4]))
print("Diferencia de Multiconjunto 1 menos Multiconjunto 2:", diferencia(mc_lista, multiconjunto([1, 2, 3, 4])))
```

La **diferencia** se refiere a los elementos que están en el primer multiconjunto pero no en el segundo. Sin embargo, a diferencia de los conjuntos normales, los multiconjuntos pueden contener múltiples instancias del mismo elemento, y estas instancias se tienen en cuenta en la operación de diferencia.

`Counter()` cuenta las ocurrencias de cada elemento en las listas. Si una cuenta en `contador1` es menor o igual a la cuenta correspondiente en `contador2`, el elemento se elimina del resultado (esto calcula la diferencia de multiconjunto). Después se obtienen los elementos del resultado con `.elements()` (convierte el resultado del contador en una lista de elementos).

Si el multiconjunto `2` tiene elementos que no están en el multiconjunto `1`, estos elementos simplemente no afectan la diferencia. La diferencia se centra en eliminar de `1` aquellos elementos que también aparecen en `2`.

```python
# Ejemplo de uso
print("Multiconjunto 1 (mc_lista):", mc_lista)
print("Multiconjunto 2:", multiconjunto([1, 1, 2, 3, 4, 6]))
print("Diferencia de Multiconjunto 1 menos Multiconjunto 2:", diferencia(mc_lista, multiconjunto([1, 2, 3, 4, 6])))
```

### Opción 2: Trabajar el multiconjunto como un Diccionario

```python
### TODO: Crear una función que dada una lista devuelva un multiconjunto
### El multiconjunto que devuelve puede crearse con la estructura de datos que se quiera (incluso una lista)

def multiconjunto(elementos):
    multiconjunto_dict = {}  # Diccionario para almacenar la frecuencia de cada elemento
    for elemento in elementos:
        if elemento in multiconjunto_dict:
            multiconjunto_dict[elemento] += 1
        else:
            multiconjunto_dict[elemento] = 1
    return multiconjunto_dict

mc = multiconjunto([1,1,1,3,3,1,4,5,1,5])
mc
```

Ahora la función `multiconjunto()` devuelve un multiconjunto representado como un **diccionario**. En este diccionario, las claves son los elementos únicos de la lista, y los valores son las frecuencias (número de veces que aparece cada elemento) en la lista. Por ejemplo, el elemento `1` aparece `5` veces.

```python
### TODO: Crear una función que dado un multiconjunto devuelva su cardinalidad

def cardinalidad(multiconjunto):
    return sum(multiconjunto.values())

print("Cardinalidad:", cardinalidad(mc))
```

La **cardinalidad** de un multiconjunto es la suma de la frecuencia de todos sus elementos. En este ejemplo, el multiconjunto tiene una cardinalidad de `10`, que se calcula sumando las frecuencias de los elementos (5 + 2 + 1 + 2).

```python
### TODO: Crear una función que dado un multiconjunto y un elemento devuelva el multiconjunto con el elemento insertado

def inserta(multiconjunto, elemento):
    if elemento in multiconjunto:
        multiconjunto[elemento] += 1
    else:
        multiconjunto[elemento] = 1
    return multiconjunto

# Ejemplo de uso
mc = multiconjunto([1,1,1,3,3,1,4,5,1,5])
mc_actualizado = inserta(mc, 9)
print("Multiconjunto actualizado: ", mc_actualizado)
print("Cardinalidad:", cardinalidad(mc_actualizado))
```

La función `inserta()` añade un elemento a un multiconjunto representado como un diccionario. Si el elemento ya existe en el multiconjunto, incrementa su contador por uno. Si el elemento no está presente, lo agrega al diccionario con un contador inicial de uno. Vamos a probar con un elemento presente en el diccionario:

```python
# Ejemplo de uso
mc = multiconjunto([1,1,1,3,3,1,4,5,1,5])
mc_actualizado = inserta(mc, 5)
print("Multiconjunto actualizado: ", mc_actualizado)
print("Cardinalidad:", cardinalidad(mc_actualizado))
```

```python
### TODO: Crear una función que dado un multiconjunto y un elemento devuelva el multiconjunto con el elemento eliminado

def elimina(multiconjunto, elemento):
    if elemento in multiconjunto:
        if multiconjunto[elemento] > 1:
            multiconjunto[elemento] -= 1
        else:
            del multiconjunto[elemento]
    else:
        print(f"El elemento {elemento} no se encuentra en el multiconjunto")
    return multiconjunto

# Ejemplo de uso
mc = multiconjunto([1,1,1,3,3,1,4,5,1,5,9])
print("Multiconjunto inicial:", mc)
mc_actualizado = elimina(mc, 9)
print("Multiconjunto después de eliminar 9:", mc_actualizado)
print("Cardinalidad:", cardinalidad(mc_actualizado))
```

```python
### TODO: Crear una función que dado un multiconjunto y un elemento devuelva si el elemento pertenece al multiconjunto

def pertenece(multiconjunto, elemento):
    respuesta = ''
    if elemento in multiconjunto:
        respuesta = f"El elemento {elemento} pertenece al multiconjunto"
    else:
        respuesta = f"El elemento {elemento} NO pertenece al multiconjunto"
    
    return respuesta

# Ejemplo de uso
print(pertenece(mc, 1))
print("Multiconjunto:", mc)
```

```python
### TODO: Crear una función que dados dos multiconjuntos devuelva si el primero es subconjunto del segundo
def subconjunto(multiconjunto1, multiconjunto2):
    for elemento, frecuencia in multiconjunto1.items():
        if elemento not in multiconjunto2 or multiconjunto2[elemento] < frecuencia:
            return False
    return True

# Ejemplo de uso
mc1 = multiconjunto([1,3,4])
mc2 = multiconjunto([1,1,1,3,3,1,4,5,1,5])
print("¿Es mc1 subconjunto de mc2?", subconjunto(mc1, mc2))  
```

Un multiconjunto es **subconjunto** de otro si cada elemento del primer multiconjunto está presente en el segundo con una frecuencia igual o mayor.

```python
# Ejemplo de uso para que devuelva False
mc3 = multiconjunto([1,1,1,1,3,3,3])
print("¿Es mc1 subconjunto de mc3?", subconjunto(mc1, mc3))
```

Dos multiconjuntos son iguales si tienen exactamente los mismos elementos con las mismas frecuencias. En este caso, la comparación de diccionarios de Python es efectiva y precisa. La comparación `multiconjunto1 == multiconjunto2` devuelve `True` si ambos multiconjuntos tienen las mismas claves con los mismos valores asociados, y `False` en caso contrario.

```python
### TODO: Crear una función que dados dos multiconjuntos devuelva su unión
def union(multiconjunto1, multiconjunto2):
    multiconjunto_union = {}

    # Añadir elementos de multiconjunto1
    for elemento, frecuencia in multiconjunto1.items():
        multiconjunto_union[elemento] = frecuencia

    # Sumar las frecuencias de los elementos de multiconjunto2
    for elemento, frecuencia in multiconjunto2.items():
        if elemento in multiconjunto_union:
            multiconjunto_union[elemento] += frecuencia
        else:
            multiconjunto_union[elemento] = frecuencia

    return multiconjunto_union
    
# Ejemplo de uso
mc1 = multiconjunto([1,3,4])
mc2 = multiconjunto([1,1,1,3,3,1,4,5,1,5])
print("Unión de mc1 y mc2:", union(mc1, mc2))
```

En la función anterior, agregamos los elementos del primer multiconjunto (`multiconjunto1`) al multiconjunto de la unión. Si un elemento no está presente (que es lo más probable, ya que el multiconjunto de la unión está vacío), se establece su frecuencia. Después sumamos las frecuencias de los elementos del segundo multiconjunto (`multiconjunto2`). Si un elemento ya está presente en el multiconjunto de la unión, incrementamos su frecuencia; de lo contrario, lo agregamos al multiconjunto con su frecuencia.

```python
# Opción 2 usando Counter
from collections import Counter

def union2(multiconjunto1, multiconjunto2):
    # Convertir los diccionarios a Contadores para facilitar la operación de unión
    contador1 = Counter(multiconjunto1)
    contador2 = Counter(multiconjunto2)

    # La unión se realiza sumando los contadores
    union_contador = contador1 + contador2
    return dict(union_contador)

# Ejemplo de uso
mc1 = multiconjunto([1,3,4])
mc2 = multiconjunto([1,1,1,3,3,1,4,5,1,5])
print("Unión de mc1 y mc2:", union2(mc1, mc2))
```

Esta función utiliza a clase `Counter()` de la biblioteca `collections` para realizar la operación de unión de manera efectiva y sencilla. La unión se realiza sumando los dos objetos `Counter`. En `Counter`, la suma combina las frecuencias de los elementos comunes y agrega los elementos que son únicos en cada multiconjunto. El resultado es un objeto `Counter`, que se convierte de nuevo a un diccionario para mantener la consistencia con la representación de los multiconjuntos de entrada.

```python
### TODO: Crear una función que dados dos multiconjuntos devuelva su intersección
def interseccion(multiconjunto1, multiconjunto2):
    multiconjunto_interseccion = {}
    
    for elemento, frecuencia in multiconjunto1.items():
        if elemento in multiconjunto2:
            multiconjunto_interseccion[elemento] = min(frecuencia, multiconjunto2[elemento])

    return multiconjunto_interseccion

# Ejemplo de uso
mc1 = multiconjunto([1,3,4])
mc2 = multiconjunto([1,1,1,3,3,1,4,5,1,5])
print("Interseccion de mc1 y mc2:", interseccion(mc1, mc2))
```

La **interseccion** de multiconjuntos resulta en un nuevo multiconjunto que contiene solo aquellos elementos que están presenten en ambos multiconjuntos originales. En el caso de multiconjuntos, donde los elementos pueden aparecer más de una vez, la intersección incluye cada elemento común con la menor frecuencia con la que aparece en cualquiera de los multiconjuntos. 

Si no hay valores coincidentes entre dos multiconjuntos, el resultado es un multiconjunto vacío (`{}`):

```python
mc3 = multiconjunto([0,6])
print("Multiconjunto 3:", mc3)
print("Multiconjunto 2:", mc2)
print("Interseccion de mc3 y mc2:", interseccion(mc3, mc2))
```

```python
### TODO: Crear una función que dados dos multiconjuntos devuelva su diferencia
def diferencia(multiconjunto1, multiconjunto2):
    multiconjunto_diferencia = {}

    for elemento, frecuencia in multiconjunto1.items():
        if elemento in multiconjunto2:
            if frecuencia > multiconjunto2[elemento]:
                multiconjunto_diferencia[elemento] = frecuencia - multiconjunto2[elemento]
        else:
            multiconjunto_diferencia[elemento] = frecuencia

    return multiconjunto_diferencia

# Ejemplo de uso
mc1 = multiconjunto([1,1,1,3,3,1,4,5,1,5])
mc2 = multiconjunto([1,2,3,4])
print("Multiconjunto 1:", mc1)
print("Multiconjunto 2:", mc2)
print("Diferencia de mc2 y mc1:", diferencia(mc1, mc2))
```

La **diferencia** entre dos multiconjuntos consiste en los elementos del primer multiconjunto que no están en el segundo multiconjunto, teniendo en cuenta las frecuencias.

En el ejemplo anterior, cuando un elemento está presente en ambos multiconjuntos y la frecuencia en `multiconjunto1` es mayor que en `multiconjunto2`, se añade este elemento al multiconjunto de la diferencia con una frecuencia igual a la diferencia de las frecuencias. Si el elemento solo está presente en `multiconjunto1`, se añade al multiconjunto de la diferencia con su frecuencia original.

```python
### TODO: Crear una función que dados dos multiconjuntos devuelva su diferencia
def diferencia2(multiconjunto1, multiconjunto2):
    multiconjunto_diferencia = {}

    for elemento, frecuencia in multiconjunto1.items():
        if elemento not in multiconjunto2:
            multiconjunto_diferencia[elemento] = frecuencia
        elif frecuencia > multiconjunto2[elemento]:
            multiconjunto_diferencia[elemento] = frecuencia - multiconjunto2[elemento]

    return multiconjunto_diferencia

# Ejemplo de uso
mc1 = multiconjunto([1,1,1,3,3,1,4,5,1,5])
mc2 = multiconjunto([1,2,3,4])
print("Multiconjunto 1:", mc1)
print("Multiconjunto 2:", mc2)
print("Diferencia de mc2 y mc1:", diferencia2(mc1, mc2))
```

Deberas crear tu propia función para calcular el error de reconstrucción, que viene definido por:

$$SSE =  \sum_{n}^{i=1}  \begin{Vmatrix}x_{i} -  \widehat{x}_i \end{Vmatrix} ^2 $$

Donde:

- $x_i$ son los valores de la matriz original X
- $\widehat{x}_i$ son los valores de la matriz reconstruida

```python
### TODO: Función para calcular el error de reconstrucción

import numpy as np

def sse_score(X, X_hat):
    return np.sum((X - X_hat)**2)
    
X = np.array([[1, 2], [3, 4]])
X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
sse = sse_score(X, X_hat)

print("Error de reconstrucción (SSE):", sse)
```

Una vez que ya tenemos la función `sse` hecha, podemos pasar a construir la función que ejecutará `SVM`.

```python
### TODO: Función para ejecutar SVM
### Tiene como entrada una matriz X
### Devuelve U, s, Vt

### Hint: S debe ser una matriz diagonal

import numpy as np

def svm(X):
    """ Función que ejecuta SVM y devuelve U, S, Vt
    
    Argumentos: X -- Matriz Original
        
    Ejemplo:
        X = np.array([[1, 2], [3, 4]])
        U, S, Vt = svm(X)  
        
        # -> U = array([[-0.40455358, -0.9145143 ],
        #               [-0.9145143 ,  0.40455358]])
        # -> S = array([[5.4649857 , 0.        ],
        #               [0.        , 0.36596619]])   
        # -> Vt = array([[-0.57604844, -0.81741556],
        #                [ 0.81741556, -0.57604844]])          
    """ 
    U, s, Vt = np.linalg.svd(X, full_matrices=False)
    S = np.diag(s) # Convertir el vector de valores singulares en una matriz diagonal
    
    # Redondeo de U, S y Vt a dos decimales
    U = np.round(U, 2)
    S = np.round(S, 2)
    Vt = np.round(Vt, 2)
    
    return U, S, Vt
    
X = np.array([[1, 2], [3, 4]])
U, S, Vt = svm(X)

print("U:", U)
print("S:", S)
print("Vt:", Vt)
```

La función de NumPy `np.linalg.svd(X, full_matrices=False)`, con el argumento `full_matrices` establecido en `False`, devuelve las matrices `U` y `Vt` en sus **formas reducidas**. En este caso, si `X` es una matriz de `m x n`, `U` será de tamaño `m x min(m, n)` y `Vt` será de tamaño `min(m, n) x n`. Esto es útil cuando trabajamos con matrices grandes, ya que reduce la cantidad de cálculos y el uso de memoria.

Como hemos visto en clase, las matrices obtenidas a partir de `SVM` nos sirven para reconstruir la matriz original `X`. Para ello, construye una función que permita reconstruir la matriz original `X` a partir de `U, s, Vt`.

```python
### TODO: Función para reconstruir la matriz original a partir de U, s, Vt
### Tiene como entrada U, s, Vt
### Devuelve X_hat

def reconstruction(U, S, Vt):
    """
    Función que reconstruye la matriz original a partir de U, s, Vt
    
    Argumentos:
        U -- Matriz de Singular Vectors
        s -- Matriz de Eigenvalues
        Vt -- Matriz de Singular Vectors
        
    Ejemplo:
        U = np.array([[-0.40455358, -0.9145143 ],
                      [-0.9145143 ,  0.40455358]])
        S = np.array([[5.4649857 , 0.        ],
                      [0.        , 0.36596619]])
        Vt = np.array([[-0.57604844, -0.81741556],
                       [ 0.81741556, -0.57604844]])
        X_hat = reconstruction(U, S, Vt)
        
        # X_hat -> array([[0.99999999, 1.99999998],
        #                 [3.00000003, 4.00000001]])
    """
    
    X_reconstruida = U.dot(S.dot(Vt))
    return X_reconstruida

# Matriz de Singular Vectors
U = np.array([[-0.40455358, -0.9145143],
              [-0.9145143,  0.40455358]])

# Matriz de Eigenvalues
S = np.array([[5.4649857 , 0.],
              [0. , 0.36596619]])

# Matriz de Singular Vectors
Vt = np.array([[-0.57604844, -0.81741556],
               [ 0.81741556, -0.57604844]])

X_hat = reconstruction(U, S, Vt)

print("Matriz reconstruida:", X_hat)
```

```python
### Probando con np.dot(U, np.dot(S, Vt)), obtenemos el mismo resultado

def reconstruction2(U, S, Vt):
    X_reconstruida = np.dot(U, np.dot(S, Vt))
    return X_reconstruida
    
# Ejemplo de uso
U = np.array([[-0.40455358, -0.9145143],
              [-0.9145143,  0.40455358]])
S = np.array([[5.4649857 , 0.],
              [0. , 0.36596619]])
Vt = np.array([[-0.57604844, -0.81741556],
               [ 0.81741556, -0.57604844]])

X_hat2 = reconstruction2(U, S, Vt)

print("Matriz reconstruida:", X_hat2)
```

```python
### TODO: Función que recibe una imagen A y devuelve la imagen comprimida
### Tiene como entrada A y el número de componentes para realizar la reducción de dimensionalidad
### Devuelve la imagen comprimidad, el error de reconstrucción y el ratio de compresión

### Hint: Usa las funciones anteriormente construidas

def image_compression(A, n_components):
    """
    Función para comprimir una imagen A
    
    Argumentos:
        A -- Imagen original
        n_components -- Número de componentes
        
    Ejemplo: A_hat, sse, comp_ratio = image_compression(A, n_components=50)
    """
    
    U, S, Vt = svm(A)

    U_reducida = U[:, :n_components] # m x r
    S_reducida = S[:n_components, :n_components] # r x r
    Vt_reducida = Vt[:n_components, :] # r x n

    # Reconstruimos matriz
    A_hat = reconstruction(U_reducida, S_reducida, Vt_reducida)

    # Error de reconstruccion
    sse = sse_score(A, A_hat)

    # Ratio de compresión m × r + r + r × n
    comp_ratio = (A.shape[0]*n_components + n_components + n_components*A.shape[1]) / (A.shape[0] * A.shape[1])

    return A_hat, sse, comp_ratio

# Ejemplo de uso
A_hat, sse, comp_ratio = image_compression(A, n_components=50)

print(A_hat)
print(sse)
print(comp_ratio)
```

Ya que conocemos la ecuación para $w_{LS}$ tenemos todo lo necesario para resolver la regresión lineal. Vamos allá!<br><br>

<center>$w_{LS} = (X^T X)^{-1}X^T y,$</center>

```python
### TODO: Función para invertir una matriz
### Contruye una función que toma como input una matriz
### Devuelve la inversa de dicha matriz

def inverse_of_matrix(mat):
    """Calcula y devuelve la inversa de la matriz
    
    Argumentos: mat -- Matriz cuadrada a invertir
        
    Ejemplo:
        sample_matrix = [[1, 2], [3, 4]]
        the_inverse = inverse_of_matrix(sample_matrix)  
        # -> the_inverse = [[-2.   1. ]
        #                   [ 1.5 -0.5]]
    
    Requerimientos: Esta función depende de 'numpy.linalg.inv'
    """
    mat = np.array(mat)
    try:
        mat_inv = np.linalg.inv(mat)
        return mat_inv
    except np.linalg.LinAlgError:
        return "La matriz no es invertible" # Manejamos el error si la matriz no es invertible

# Ejemplo de uso
sample_matrix = [[1, 2], [3, 4]]
the_inverse = inverse_of_matrix(sample_matrix) 

print("Matriz original:\n", sample_matrix)
print("Matriz inversa:\n", the_inverse)
```

#### Mínimos Cuadrados / Least Squares

Ahora, implementarás la ecuación $w_{LS}$:

<center>$w_{LS} = (X^T X)^{−1}X^T y,$</center>

```python
### TODO: Función para resolver la ecuación wLS
### Toma como argumentos dos matrices, una para X y otra para y
### Asumimos que las matrices tienen las dimensiones correctas

### Paso 1: Asegurate que n > d. 
### Es decir, que el número de observaciones es mayor que el número de dimensiones.
### O lo que es lo mismo, que el número de filas de cada matriz sea mayor que el número de columnas
### Si no es así, debes transponer las matrices

### Paso 2: Debes añadir a la matriz X un vector columna del tamaño (n x 1)

### Paso 3: Usa la ecuación de arriba para obtener wLS

import numpy as np

def least_squares_weights(input_x, target_y):
    """Resuelve la ecuación para wLS
    
    Argumentos:
        input_x -- Matriz con los datos de entrenamiento
        target_y -- Vector con los datos de salida

    Asumimos: target_y es un vector con el mismo número de observaciones que input_x
    """
    # Asegurar primero que n (nº observaciones) > d (nº dimensiones o caracteristicas)
    # Se espera que input_x sea una matrix con dimensiones n x d (n filas, d columnas)
    # Cada fila de input_x representa una observación individual y cada columna una característica diferente
    # Si no es así, se transpone tanto input_x como target_y:
    if input_x.shape[0] < input_x.shape[1]:
        # La transposicion convierte la matrix de d x n a n x d
        input_x = input_x.T
        target_y = target_y.T
    
    # Añadimos un vector columna de unos a la matriz X
    ones = np.ones((input_x.shape[0], 1))
    X = np.hstack((ones, input_x))
    
    # Resolvemos la ecuación de mínimos cuadrados
    X_t = np.transpose(X)
    prod = np.dot(X_t, X)
    inv_prod = np.linalg.inv(prod)
    wLS = inv_prod @ X_t @ target_y
    
    return wLS
    

# Ejemplo de uso
training_y = np.array([[208500, 181500, 223500, 140000, 250000,
                        143000, 307000, 200000, 129900, 118000]])
training_x = np.array([[1710, 1262, 1786, 1717, 2198,
                        1362, 1694, 2090, 1774, 1077], 
                        [2003, 1976, 2001, 1915, 2000,
                         1993, 2004, 1973, 1931, 1939]])

weights = least_squares_weights(training_x, training_y)

# Imprimir los pesos obtenidos
print(weights)  # --> np.array([[-2.29223802e+06],
                #              [ 5.92536529e+01],
                #              [ 1.20780450e+03]])

# Imprimir el primer coeficiente (excluyendo el término constante)
print(weights[1][0]) # --> 59.25365290008861
```

## 4. Linear Regression - Gradient Descent

En este ejercicio resolveras el mismo problema anterior pero usando **Gradient Descent**

**Objetivos**:
- Asegurar los fundamentos matemáticos detrás del Gradient Descent.

**Problema**: Usando datos sobre el precio de la vivienda, intentaremos predecir el precio de una casa en base a la superficie habitable con un modelo de regresión.

**Datos:** [Kaggle's House Prices Dataset](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data)

**Repaso:**

$$ RSS(w) = \sum_{n=1}^{N}[y_n-f(x_n)]^2 =  \sum_{n=1}^{N}[y_n- (w_0 + \sum_{d=1}^{D}w_dx_{nd}) ]^2 .$$

Loss function:

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

Y lo que queremos es minimizar esta distancia, para que el modelo se acerque lo máximo posible a los valores verdaderos.

$$\nabla RSS(w) = X^T(Xw^t-y)$$

En resumen, el gradient descendiente para una regresión lineal, se basa en resolver esta ecuación de forma iterativa:

$$w^{t+1} = w^t - \eta * \nabla RSS(w)$$

#### Leer Datos

```python
import pandas as pd
import numpy as np

# Leer datos
data = pd.read_csv('./data/train.csv')

# Extraer dichas columnas
newData = data[['GrLivArea','SalePrice']]
print(newData.head())

# Construir x - y
x = newData['GrLivArea']
y = newData['SalePrice']

# Standarizar los datos
x = (x - x.mean()) / x.std()
x = np.c_[np.ones(x.shape[0]), x] 

print("Shape of X: ", x.shape)
print("Shape of y:", y.shape)
```

#### Gradient Descent

```python
### TODO: Función para encontrar los valores w usando Gradient Descent
### Toma como argumentos: X, y, w, n_iterations, eta
### Completa la función añadiendo la loss función y la updating rule

def gradient_descent(x, y, w, iterations, eta):
    
    num_obs = y.size  # Número de observaciones
    loss_list = [] # Lista para almacenar la pérdida (RSS) de cada iteración
    
    # Matriz para almacenar los pesos en cada iteración
    weights = np.zeros((iterations, w.size))  # Tiene tantas filas como iteraciones y tantas columnas como elementos en w

    for i in range(iterations):
        
        # Predicciones para la hipótesis actual
        predictions = x.dot(w)  # XW
        
        # Error (diferencia entre predicciones y valores reales)
        error = y - predictions   # y - XW

        # Cálculo de la pérdida (Residual Sum of Squares, RSS)
        loss = (1/2) * np.sum(error**2)  # ((1/2) * (y - predictions)**2)
        loss_list.append(loss)

        # Gradiente
        gradient = x.T.dot(x.dot(w) - y)

        # Actualización de los pesos
        w = w - eta * gradient
        weights[i] = w  # Almacenar los pesos actuales

    return weights, loss_list
    
    
# Learning rate
eta = 0.01 

# Número de iteraciones
iterations = 2000 #No. of iterations

# Seed para inicializar w
np.random.seed(123)
w0 = np.random.rand(2) # Vector aleatoriamente inicializado

# Matriz con los datos de entrenamiento
training_x = np.array([[ 1.        ,  0.37020659],
                       [ 1.        , -0.48234664],
                       [ 1.        ,  0.51483616],
                       [ 1.        ,  0.38352774],
                       [ 1.        ,  1.29888065]])

# Vector con los datos de salida
training_y = np.array([208500, 181500, 223500, 
                        140000, 250000])

weights, loss = gradient_descent(training_x, training_y, w0, iterations, eta)

print(weights[-1]) #--> np.array([183845.82320222  40415.66453324])
print("Pérdida en la última iteracion:", loss[-1])
```

**Pesos después del Descenso de Gradiente:**

- El primer valor (`183845.82320222`) es el término de intercepción (término constante o bias). Representa el precio estimado de la vivienda cuando la variable independiente (superficie habitable) es cero. Dado que no es realista tener una vivienda con superficie cero, este valor puede interpretarse más como un ajuste que ayuda a calibrar el modelo.

- El segundo valor (`40415.66453324`) es el coeficiente asociado con la variable independiente, la superficie habitable. Indica cuánto se espera que cambie el precio de la vivienda por cada unidad de cambio en la superficie habitable. Por ejemplo, si la superficie habitable aumenta en una unidad, el precio de la vivienda aumentaría, en promedio, en aproximadamente 40,415 unidades monetarias.

Una vez construida nuestra función para el Gradient Descent podemos usarla para encontrar los valores optimos de $w$. **Prueba a modificar el learning rate para ver la convergencia del Gradient Descent.**

```python
import re
import lxml
from bs4 import BeautifulSoup
import urllib
import urllib.request

url = "https://es.wikipedia.org/wiki/Canis_lupus_familiaris"

def parse_from_url(url):
    """
    Función para extraer el contenido (raw text) de una página web
    """
    
    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, "html.parser" )
    for script in soup(["script", "style"]):
        script.extract()
        
    text = soup.get_text()
    
    # Eliminar saltos de linea
    text = re.sub('\s+', ' ', text)
    return text

parse_from_url(url)[:1000]
```

```python
### TODO: Escribe una función que recibe una lista de N páginas web
### Y calcula todas las combinaciones posibles entre ellas, no importa el orden
### [web_1, web_2, web_3, ...]
### Devuelve una lista de tuples con las combinaciones [(web_1, web_2), (web_1, web_3), ...]

# HINT: Puedes implementar esta función como quieras pero la librería itertools 
#       proporciona una función llamada `combinations` para realizar esta tarea.

import itertools
    
def combinations(url_ls):
    # Calcula todas las combinaciones posibles de dos elementos de la lista url_ls
    # itertools.combinations(iterable, r) crea combinaciones de longitud r
    return list(itertools.combinations(url_ls, 2))
       
url_ls = ['https://es.wikipedia.org/wiki/Canis_lupus_familiaris', 
        'https://es.wikipedia.org/wiki/Canis_lupus',
        'https://es.wikipedia.org/wiki/Felis_silvestris_catus']
        
permutation = combinations(url_ls) 
print("Combinaciones posibles:",permutation)
print("Número de combinaciones posibles:",len(permutation))
```

```python
### TODO: Escribe una función que recibe una lista con tuples
### [({'que', 'hola'}, {'que', 'es', 'guay'}), ({'que', 'hola'}, {'madrid', 'la', 'es'})]
### Y devuelve una lista con la union de los conjuntos
### [({'que', 'hola', 'es', 'guay'}), ({'que', 'hola', 'madrid', 'la', 'es'})]

def union(comb_ls):
    final_list = []
    for tuple_set in comb_ls:
        union = set()
        for item in tuple_set:
            for word in item:
                union.add(word)
        final_list.append(union)
    return final_list
    
# Ejemplo de uso
comb_ls = [({'que', 'hola'}, {'que', 'es', 'guay'}), ({'que', 'hola'}, {'madrid', 'la', 'es'})]
union_ls = union(comb_ls)  # -> [{'es', 'que', 'hola', 'guay'}, {'es', 'que', 'hola', 'madrid', 'la'}]
print(union_ls)
```

Una vez que tenemos una lista de conjuntos por cada par de páginas web, podemos convertir el texto de la página web a un vector.

```python
def set2vector(tokens_web1, tokens_web_2, set_web1, set_web2):
    """
    Función para convertir un conjunto a vector
    Requerimientos: Depende de la función `union()`
    """
    
    # Unimos los conjuntos
    join_set = union([(set_web1, set_web2)])[0]
    
    web1_array = []
    web2_array = [] 

    for word in join_set:
        if word in tokens_web1:
            web1_array.append(1)
        else:
            web1_array.append(0)
        if word in tokens_web_2:
            web2_array.append(1)
        else:
            web2_array.append(0)

    return web1_array, web2_array

# Contenido tokenizado de página web 1
tokens_web1 = ["hola", "que", "tal", "soy", "Rocío"]

# Contenido tokenizado de página web 2
tokens_web_2 = ["hola", "me", "llamo"]

# Conjunto de palabras de la página web 1
set_web1 = {"hola", "que", "tal", "soy", "Rocío"}

# Conjunto de palabras de la página web 2
set_web2 = {"hola", "me", "llamo"}

web1_array, web2_array = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2) 
```

### Cosine Similarity

Por último, ya podemos implementar la ecuación: $\cos \theta = \frac{\overrightarrow{u} \cdot \overrightarrow{v}}{|\overrightarrow{u}||\overrightarrow{v}|}$

```python
### TODO: Escribe una función que recibe dos vectores, u y v
### Y devuelva la similaridad entre ambos vectores
### Paso 1: Si u y v son listas -> Convertirlo a arrays
### Paso 2: Calcula la similaridad entre ambos vectores

import numpy as np

def cosine_similarity(u, v):
    if type(u) == list:
        u = np.array(u)
    if type(v) == list:
        v = np.array(v)
    similarity = (u @ v) / (np.linalg.norm(u) * np.linalg.norm(v))
    return similarity

# Ejemplo de uso
u = np.array([1, 2, 3])
v = np.array([3, 2, 1])

similarity = cosine_similarity(u, v)  # -> 0.71428
print(similarity)
```
